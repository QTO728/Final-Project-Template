For this first graph, i used the mussel length data and the salinity level data to graph the relationship. This will help answer my question of how salinity level affect mussel length. I intend to create a twin graph with the error bar showing the high and low of mussel length throughout the year (2014-2015) and the salinity level to be graphed on top of it. 

```python
# load data file(s)
first_source = "mussel_morphologyBioenergetics_kenaiFjords_robinson_2014-2015 (2).csv"
third_source = "UAF_GAK1_2014_ctd014m_L2_v2.csv"

mussel_df = pd.read_csv(first_source, delimiter=',', delim_whitespace=False,
                       header=0, parse_dates=['CollectionDate'])
salinity_df = pd.read_csv(third_source, parse_dates=['Date_Time_[UTC]'])
# as part of commenting your code throughout this section, discuss any problems you encountered and how you solved them
```

```python
# perform data cleaning steps
mussel_df = mussel_df.sort_values("CollectionDate")

# Set the time as the index in both data
mussel_df.set_index('CollectionDate', inplace=True)
salinity_df.set_index('Date_Time_[UTC]', inplace = True)

# Groups the data into max and min data.
mussel_max = mussel_df.groupby(mussel_df.index).max(numeric_only = True)
mussel_min = mussel_df.groupby(mussel_df.index).min(numeric_only =True)
```

```python
# perform data analysis
mussel_df = mussel_df['Length'].resample('M').mean()
mussel_df_sel = mussel_df.loc[:'2015-03-01']
mussel_df_sel = mussel_df_sel.interpolate()

mussel_max = mussel_max['Length'].resample('M').mean()
mussel_min = mussel_min['Length'].resample('M').mean()

# Index mussel data to the wanter time range
mussel_max_sel = mussel_max.loc[:'2015-03-01']
mussel_min_sel = mussel_min.loc[:'2015-03-01']

# Gets the absolute values for the error bar.
max_diff = np.abs(mussel_df_sel - mussel_max_sel)
min_diff = np.abs(mussel_df_sel - mussel_min_sel)

# Creates 2D array for the error bar
errs = np.empty((2, len(max_diff)))
errs[0] = min_diff
errs[1] = max_diff

# Gets the monthly mean of the salinity level.
salinity_df = salinity_df['Salinity_[psu]'].resample('M').mean()
# Index the data to this time range.
salinity_df_sel = salinity_df.loc['2014-04-30':'2015-03-01']

# Some limitations are that data has to be sorted according to time, and some month has lower 
# data and some months has a lot more data, so it was hard to show all the data points on
# the graph.
```

```python
# make and display the first figure
# Plots figures
fig,ax = plt.subplots(figsize=(10,5))
line1 = ax.errorbar(mussel_df_sel.index, mussel_df_sel, yerr=errs, fmt='.-', 
                    markersize=15, capsize=5, capthick=2, c='darkblue', label="Length")
ax.set_xlabel('Time (month)')
ax.set_ylabel('Length (mm)')
# Rotate the time on the x-axis
plt.xticks(rotation=45)

# Creates a twin graph
ax2 = ax.twinx()
ax2.plot(salinity_df_sel.index, salinity_df_sel, c='g', linestyle='-', 
         label="Salinity", marker='o', markersize=6)

# Adds legends
lines, labels = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax.legend(lines + lines2, labels + labels2, loc="lower right")
#ax.legend(loc="lower right")
ax2.set_ylabel('Salinity (psu)')
ax.grid()
ax.set_title("Mussel's Length in Kenai Fjords National Park (2014-2015)")
```

<!-- Describe how this figure helps address your research question. What patterns or insights can you observe? -->
**Figure 1 Caption:** Your caption here.
This graph shows the salinity level decreased from 05/2014 - 10/2014 and increased from 10/2014 - 03/2015. The error bars show the range of the length, with longest and shortest length and the mean in the middle. There is kind of a trend when the length started to increased from 10/2014 - 03/2015 as salinity also increased. The mussel's length mean also highest around 02/2015 and 03/2015. 

## 2. {Section 2 Title}  

<!-- In the introduction below, specify the data set(s) you are working with in this section, which aspect of your research question this section addresses, and the figure you will produce -->

Short introduction to this section here.
The data i use for this second graph is the mussel length data and the water temperature data from the 3rd source. The figure i will produce is a scatter plot with a color bar, the plot will shows the length in that month with the fitting color represents the water temperature in that month. 

```python
# if a new dataset, load data and perform data cleaning steps
# otherwise, continue to data analysis and figure creation
# Data imports
sec_source = "Intertidal_Temperature_KEFJ_2014-2015.csv"
# Reads in the file and set the time to datetime object.
water_df = pd.read_csv(sec_source, delimiter=',', delim_whitespace=False,
                       header=0, parse_dates=['date'])

# Set time as index
water_df.set_index('date', inplace = True)
```

```python
# perform data analysis
# Resample the data to monthly mean
water_temp = water_df['temperature'].resample('M').mean()

# Index the data to the desired time range
water_temp_sel = water_temp.loc['2014-06-01':'2015-03-01']
mussel_df_sel2 = mussel_df.loc['2014-06-01':'2015-03-01']
# Interpolates the missing data from mussel data.
mussel_df_sel2 = mussel_df_sel2.interpolate()

# A limitation i found were that the time scale of the water temperature data is really short, 
# only 9 months, which is hard to plot with the mussel data. Also i don't know about 
# depth that the mussel data were collected so the data accuracy is not good. 
```

```python
# make and display figure
fig1,ax1 = plt.subplots(figsize=(10,5))
sc = ax1.scatter(mussel_df_sel2.index, mussel_df_sel2, c=water_temp_sel)
# makes a color bar
cbar = fig1.colorbar(sc)
cbar.set_label("Temperature (C)", fontsize=10)
ax1.set_xlabel('Time (months)')
ax1.set_ylabel('Length (mm)')
ax1.set_title("Mussel's Length vs. water temperature")
# rotates the time on x-axis
plt.xticks(rotation=45)
ax1.grid()

```

<!-- Describe how this figure helps address your research question. What patterns or insights can you observe? -->
**Figure 2 Caption:** 
This graph shows the temperature and mussel length from 07/2014 - 03/2015. There is a trend of temperature decreasing as the graph shifts into winter season from summer. There is a drop in mussel length on October 2014. The mussel length increases with the temperature decreases. The highest mussel length is 37 mm on 02/2015.
